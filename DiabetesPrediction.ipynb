#### Importing necessary libraries

```python
import pandas as pd                 #Import Pandas library for data manipulation and analysis 
import seaborn as sns               #Import Seaborn library for statiscal data visualization
import matplotlib.pyplot as plt     #Import Matplotlib library for data visualization
from sklearn.model_selection import train_test_split   #Import train_test_split function for splittng data
from sklearn.linear_model import LogisticRegression    #Import LogisticRegression class from logistic regression
from sklearn.metrics import classification_report     #Import classification_report function for detailed classification metrics
from sklearn.metrics import f1_score , accuracy_score  #Import accuracy_score to calculate the accuracy of a model
```

```python
df=pd.read_csv('diabetes (1).csv')
```

We can observe that the min value of some column appears to be 0 which cannot be medically possible.Hence, in the data cleaning process we have to replace them with their mean/median values

##### Cleaning the data

```python
#Checking if any null values present
df.isnull().sum()
```

```python
#Replacing  0 values with mean of its column having noraml distribution

df['Glucose']=df['Glucose'].replace(0,df['Glucose'].mean())
df['BloodPressure']=df['BloodPressure'].replace(0,df['BloodPressure'].mean())

#Replacing  0 values with median of its column having skewed distribution

df['SkinThickness']=df['SkinThickness'].replace(0,df['SkinThickness'].median())
df['Insulin']=df['Insulin'].replace(0,df['Insulin'].median())
df['BMI']=df['BMI'].replace(0,df['BMI'].median())
```

##### Data visualization

```python
#sns.countplot('Outcome',data=df)
```

Number of people who do not have diabetes is more than that of the people having diabetes, this indicates imbalance in data

```python
df.hist(bins=10,figsize=(8,8))
plt.show()
```

We observe that only Glucose and BloodPressure are normally distributed rest others are skewed distribution

```python
corr_mat=df.corr()
sns.heatmap(corr_mat , annot= True)
```

We can observe that Glucose,BMI,Age  are most correlated with Outcome whereas BloodPressure,Insulin,DiabetesPedigreeFunction are the least correlated  and so we can drop them

```python
df=df.drop(['BloodPressure','Insulin','DiabetesPedigreeFunction'],axis ='columns')
```

```python
df
```

#### Splitting data into training and testing set

```python
#Defining feature variable X and target variable y
X=df.drop('Outcome',axis=1)
y=df['Outcome']
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y ,test_size =0.2 )

print(X_train.shape,y_train.shape)
print(X_test.shape, y_test.shape)
```

###### Create and train the model

```python
model=LogisticRegression()
model.fit(X_train , y_train)
```

```python
#Generating Predictions
y_predicted=model.predict(X_test)
```

#### Evaluate the model

```python
print(classification_report(y_test ,y_predicted))
```

```python
print(f1_score(y_test , y_predicted))
```

```python
print(accuracy_score(y_test , y_predicted))
```
